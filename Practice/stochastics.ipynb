# Simulation
Monte Carlo simulation (MCS) is among the most important numerical techniques in finance, if not the most important and widely used. This mainly stems from the fact that it is the most flexible numerical method when it comes to the evaluation of mathematical expressions (e.g., integrals), and specifically the valuation of financial derivatives.

```python

# Initial stock price
S0 = 100
# Risk-free rate
r = 0.05
# Volatility
sigma = 0.25
# Time horizon
T = 2.0
# Number of simulations
I = 10000

# Simulate end-of-period values for the stock price 
# using geometric Brownian motion
ST1 = S0 * np.exp((r - 0.5 * sigma ** 2) * T +
                  sigma * math.sqrt(T) * npr.standard_normal(I))

# Plot the simulated end-of-period stock prices
plt.figure(figsize=(10, 6))
plt.hist(ST1, bins=50)
plt.xlabel('index level')
plt.ylabel('frequency')
plt.show()

#Statiscially simlated geometric Brownian motion.
```

```python
import numpy as np
import scipy.stats as scs

def print_statistics(a1, a2):
    '''
    Prints selected statistics.

    Parameters
    ==========
    a1, a2: ndarray objects
        results objects from simulation
    '''
    # Get descriptive statistics for both datasets
    sta1 = scs.describe(a1)
    sta2 = scs.describe(a2)

    # Print header
    print('%14s %14s %14s' % ('statistic', 'data set 1', 'data set 2'))
    print(45 * "-")

    # Print statistics
    print('%14s %14.3f %14.3f' % ('size', sta1[0], sta2[0]))
    print('%14s %14.3f %14.3f' % ('min', sta1[1][0], sta2[1][0]))
    print('%14s %14.3f %14.3f' % ('max', sta1[1][1], sta2[1][1]))
    print('%14s %14.3f %14.3f' % ('mean', sta1[2], sta2[2]))
    print('%14s %14.3f %14.3f' % ('std', np.sqrt(sta1[3]), np.sqrt(sta2[3])))
    print('%14s %14.3f %14.3f' % ('skew', sta1[4], sta2[4]))
    print('%14s %14.3f %14.3f' % ('kurtosis', sta1[5], sta2[5]))

# Call the function with your data
print_statistics(ST1, ST2)
```

# Square Root Diffusion

The notebook acknowledges that \(x_t\) values follow a chi-squared distribution. Many financial models can be approximated with a normal distribution using an Euler discretization scheme. While accurate for geometric Brownian motion, the Euler scheme introduces bias for other stochastic processes. An Euler scheme, known as full truncation, is presented as \(x_t\) and \(z_t\) in Equation 12-5. This approach is preferred for numerical or computational reasons, even if an exact scheme is available for the square-root diffusion, which will be discussed later.

**Euler Discretization for Square-Root Diffusion (Equation 12-5):**

The square-root diffusion ensures \(x_t\) remains positive. When discretized with an Euler scheme, negative values may arise. Thus, working with the positive version of the simulated process requires two ndarray objects in the simulation code. The results are depicted in histogram form, as shown in Figure 12-7.


```python


# Initial value
x0 = 0.05
# Speed of mean reversion
kappa = 3.0
# Long-term mean of the process
theta = 0.02
# Volatility
sigma = 0.1
# Number of simulations
I = 10000
# Number of time steps
M = 50
# Time step size
dt = T / M

def srd_euler():
    '''
    Simulates a square-root diffusion process using the Euler method.
    '''
    # Initialize arrays to store simulations
    xh = np.zeros((M + 1, I))
    x = np.zeros_like(xh)
    # Set initial values
    xh[0] = x0
    x[0] = x0

    # Simulate paths
    for t in range(1, M + 1):
        xh[t] = (xh[t - 1] +
                 kappa * (theta - np.maximum(xh[t - 1], 0)) * dt +
                 sigma * np.sqrt(np.maximum(xh[t - 1], 0)) *
                 math.sqrt(dt) * npr.standard_normal(I))
    x = np.maximum(xh, 0)
    return x

# Run the simulation
x1 = srd_euler()

# Plot the final distribution of simulations
plt.figure(figsize=(10, 6))
plt.hist(x1[-1], bins=50)
plt.xlabel('value')
plt.ylabel('frequency')
plt.show()
```

## Exact discretization for square-root diffusion


```python

def srd_exact():
    '''
    Simulates a square-root diffusion process using exact discretization.
    '''
    # Initialize array to store simulations
    x = np.zeros((M + 1, I))
    # Set initial value
    x[0] = x0

    # Simulate paths
    for t in range(1, M + 1):
        # Degrees of freedom of non-central chi-square distribution
        df = 4 * theta * kappa / sigma ** 2
        # Scale parameter of non-central chi-square distribution
        c = (sigma ** 2 * (1 - np.exp(-kappa * dt))) / (4 * kappa)
        # Non-centrality parameter of non-central chi-square distribution
        nc = np.exp(-kappa * dt) / c * x[t - 1]
        # Draw random numbers from non-central chi-square distribution
        x[t] = c * npr.noncentral_chisquare(df, nc, size=I)
    return x

# Run the simulation
x2 = srd_exact()

# Plot the final distribution of simulations
plt.figure(figsize=(10, 6))
plt.hist(x2[-1], bins=50)
plt.xlabel('value')
plt.ylabel('frequency')
plt.show()
```

- **Description:**

- **Simplicity:**
  - Euler discretization is straightforward to implement, making it a popular choice for numerical simulations.

- **Approximation:**
  - It involves approximating the continuous-time stochastic process by taking small steps forward in time. However, it introduces bias, especially for certain types of stochastic processes.

- **Use Cases:**
  - *Example:* When computational efficiency is a primary consideration, and an exact scheme is not necessary. Euler discretization is commonly used in options pricing models like Black-Scholes.

## Exact Discretization for Square-Root Diffusion (Equation 12-6):

- **Description:**
  - The "Equation 12-6" represents an exact discretization method specifically designed for the square-root diffusion.

- **Accuracy:**
  - Unlike the Euler scheme, the exact discretization aims to provide a more accurate representation of the dynamics of the square-root diffusion.

- **Properties:**
  - This method considers the characteristics of the square-root diffusion, ensuring that the discretization maintains specific properties of the original continuous-time process.

- **Positivity Constraint:**
  - Notably, the exact discretization for square-root diffusion explicitly addresses the positivity constraint, ensuring that simulated values remain strictly positive.

- **Use Cases:**
  - *Example:* When modeling processes where strict positivity is crucial, such as interest rate models. Exact discretization is suitable for situations where maintaining specific properties of the continuous process is essential.

## Considerations:

# Stochastic Volatility


```python
import numpy as np

# Initial stock price
S0 = 100.
# Risk-free rate
r = 0.05
# Initial volatility
v0 = 0.1
# Speed of mean reversion
kappa = 3.0
# Long-term mean of volatility
theta = 0.25
# Volatility of volatility
sigma = 0.1
# Correlation between stock price and volatility
rho = 0.6
# Time horizon
T = 1.0

# Create correlation matrix
corr_mat = np.zeros((2, 2))
corr_mat[0, :] = [1.0, rho]
corr_mat[1, :] = [rho, 1.0]

# Compute Cholesky decomposition of correlation matrix
cho_mat = np.linalg.cholesky(corr_mat)

# Print Cholesky matrix
print(cho_mat)
```

```python
import numpy as np
import numpy.random as npr
import math

# Number of time steps
M = 50
# Number of simulations
I = 10000
# Time step size
dt = T / M

# Generate standard normal random numbers
ran_num = npr.standard_normal((2, M + 1, I))

# Initialize arrays to store simulations of volatility
v = np.zeros_like(ran_num[0])
vh = np.zeros_like(v)

# Set initial values of volatility
v[0] = v0
vh[0] = v0

# Simulate volatility paths
for t in range(1, M + 1):
    # Generate correlated random numbers
    ran = np.dot(cho_mat, ran_num[:, t, :])
    # Update volatility using Euler method
    vh[t] = (vh[t - 1] +
             kappa * (theta - np.maximum(vh[t - 1], 0)) * dt +
             sigma * np.sqrt(np.maximum(vh[t - 1], 0)) *
             math.sqrt(dt) * ran[1])

# Ensure volatility is non-negative
v = np.maximum(vh, 0)
```

```python
import numpy as np
import numpy.random as npr
import math
import matplotlib.pyplot as plt

# Initialize array to store simulations of stock price
S = np.zeros_like(ran_num[0])
# Set initial stock price
S[0] = S0

# Simulate stock price paths
for t in range(1, M + 1):
    # Generate correlated random numbers
    ran = np.dot(cho_mat, ran_num[:, t, :])
    # Update stock price
    S[t] = S[t - 1] * np.exp((r - 0.5 * v[t]) * dt +
                             np.sqrt(v[t]) * ran[0] * np.sqrt(dt))

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6))

# Plot histogram of final stock prices
ax1.hist(S[-1], bins=50)
ax1.set_xlabel('index level')
ax1.set_ylabel('frequency')

# Plot histogram of final volatilities
ax2.hist(v[-1], bins=50)
ax2.set_xlabel('volatility')

# Show plots
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10, 6))

# Plot the first 10 paths of stock prices
ax1.plot(S[:, :10], lw=1.5)
ax1.set_ylabel('index level')

# Plot the first 10 paths of volatilities
ax2.plot(v[:, :10], lw=1.5)
ax2.set_xlabel('time')
ax2.set_ylabel('volatility')

# Show plots
plt.show()
```

# Jump Diffusion

Stochastic volatility and the leverage effect are observed patterns, referred to as stylized (empirical) facts, found in various financial markets. Another significant stylized fact is the presence of jumps in asset prices and, for instance, volatility. In 1976, Merton introduced his jump diffusion model, which extends the Black-Scholes-Merton framework by incorporating a model component that generates jumps with a log-normal distribution. The risk-neutral Stochastic Differential Equation (SDE) capturing these features is presented in the following.

Stocahstic diffeential equation for Merton jump diffusion


Euler Discretization for MErton jup diffusion model



```python
import math

# Initial stock price
S0 = 100.
# Risk-free rate
r = 0.05
# Volatility
sigma = 0.2
# Jump intensity
lamb = 0.75
# Expected jump size
mu = -0.6
# Standard deviation of jump
delta = 0.25
# Adjusted drift to account for jump
rj = lamb * (math.exp(mu + 0.5 * delta ** 2) - 1)

# Time horizon
T = 1.0
# Number of time steps
M = 50
# Number of simulations
I = 10000
# Time step size
dt = T / M
```

```python
import numpy as np
import numpy.random as npr
import math
import matplotlib.pyplot as plt

# Initialize array to store simulations of stock price
S = np.zeros((M + 1, I))
# Set initial stock price
S[0] = S0

# Generate standard normal random numbers for diffusion and jump component
sn1 = npr.standard_normal((M + 1, I))
sn2 = npr.standard_normal((M + 1, I))

# Generate Poisson distributed random numbers for jump component
poi = npr.poisson(lamb * dt, (M + 1, I))

# Simulate stock price paths
for t in range(1, M + 1, 1):
    # Update stock price with diffusion and jump component
    S[t] = S[t - 1] * (np.exp((r - rj - 0.5 * sigma ** 2) * dt +
                               sigma * math.sqrt(dt) * sn1[t]) +
                        (np.exp(mu + delta * sn2[t]) - 1) *
                        poi[t])
    # Ensure stock price is non-negative
    S[t] = np.maximum(S[t], 0)

# Create figure
plt.figure(figsize=(10, 6))

# Plot histogram of final stock prices
plt.hist(S[-1], bins=50)
plt.xlabel('value')
plt.ylabel('frequency')

# Show plot
plt.show()
```

# Generation Function 

This function creates random numbers for Monte Carlo simulation. It accepts inputs such as the number of time intervals (M) for discretization, the number of paths (I) to be simulated, and two boolean flags (anti_paths and mo_match) indicating whether to utilize antithetic variates and moment matching, respectively.

Antithetic variates is a variance reduction technique where each random number is paired with its negative, effectively creating a mirrored image and reducing the variance of the estimate.

Moment matching is a technique in which the random numbers are standardized to have a mean of 0 and a standard deviation of 1. This is achieved by subtracting the mean and dividing by the standard deviation.

```python


def gen_sn(M, I, anti_paths=True, mo_match=True):
    ''' 
    Function to generate random numbers for simulation.

    Parameters
    ==========
    M: int
        number of time intervals for discretization
    I: int
        number of paths to be simulated
    anti_paths: boolean
        use of antithetic variates
    mo_match: boolean
        use of moment matching
    '''
    # If antithetic variates are to be used
    if anti_paths is True:
        # Generate half the random numbers
        sn = npr.standard_normal((M + 1, int(I / 2)))
        # Concatenate with their negatives
        sn = np.concatenate((sn, -sn), axis=1)
    else:
        # Generate all random numbers
        sn = npr.standard_normal((M + 1, I))
    
    # If moment matching is to be used
    if mo_match is True:
        # Standardize the random numbers
        sn = (sn - sn.mean()) / sn.std()
    
    return sn
```

# Valuations

Monte Carlo simulation finds crucial application in the valuation of contingent claims, encompassing options, derivatives, hybrid instruments, and more. Put simply, within a risk-neutral environment, the value of a contingent claim equates to the discounted expected payoff under the risk-neutral (martingale) measure. This particular probability measure ensures that all risk factors, such as stocks and indices, drift at the riskless short rate, rendering the discounted processes martingales. The Fundamental Theorem of Asset Pricing establishes that the existence of such a probability measure is synonymous with the absence of arbitrage.

A financial option grants the holder the right to either buy (call option) or sell (put option) a specified financial instrument at a predetermined maturity date (European option) or over a specified timeframe (American option), all at a given price known as the strike price. Let's begin by examining the valuation of European options for a more straightforward understanding.

# European Call Options 


Pricing by risk neutral expectation


Risk neutral Monte Carlo estimator



```python
import numpy as np
import math

# Define initial stock price, risk-free rate, volatility, time horizon, and number of simulations
S0 = 100.
r = 0.05
sigma = 0.25
T = 1.0
I = 50000

def gbm_mcs_stat(K):
    ''' 
    Valuation of European call option in Black-Scholes-Merton
    by Monte Carlo simulation (of index level at maturity)

    Parameters
    ==========
    K: float
        (positive) strike price of the option

    Returns
    =======
    C0: float
        estimated present value of European call option
    '''
    # Generate random numbers
    sn = gen_sn(1, I)
    
    # Simulate index level at maturity
    ST = S0 * np.exp((r - 0.5 * sigma ** 2) * T + sigma * math.sqrt(T) * sn[1])
    
    # Calculate payoff at maturity
    hT = np.maximum(ST - K, 0)
    
    # Calculate MCS estimator
    C0 = math.exp(-r * T) * np.mean(hT)
    
    return C0

# Call the function with a strike price of 105
print(gbm_mcs_stat(K=105.))  # Output: 10.044221852841922
```

```python
import numpy as np
import math

# Define the number of time intervals for discretization
M = 50

def gbm_mcs_dyna(K, option='call'):
    ''' 
    Valuation of European options in Black-Scholes-Merton
    by Monte Carlo simulation (of index level paths)

    Parameters
    ==========
    K: float
        (positive) strike price of the option
    option : string
        type of the option to be valued ('call', 'put')

    Returns
    =======
    C0: float
        estimated present value of European call option
    '''
    # Calculate the time interval
    dt = T / M
    
    # Initialize the array to store the index level paths
    S = np.zeros((M + 1, I))
    S[0] = S0
    
    # Generate random numbers
    sn = gen_sn(M, I)
    
    # Simulate index level paths
    for t in range(1, M + 1):
        S[t] = S[t - 1] * np.exp((r - 0.5 * sigma ** 2) * dt + sigma * math.sqrt(dt) * sn[t])
    
    # Calculate payoff based on the type of option
    if option == 'call':
        hT = np.maximum(S[-1] - K, 0)
    else:
        hT = np.maximum(K - S[-1], 0)
    
    # Calculate MCS estimator
    C0 = math.exp(-r * T) * np.mean(hT)
    
    return C0

# Call the function with a strike price of 110 for a call option
print(gbm_mcs_dyna(K=110., option='call'))  # Output: 7.950008525028434

# Call the function with a strike price of 110 for a put option
print(gbm_mcs_dyna(K=110., option='put'))  # Output: 12.629934942682004
```

# Static Simulation (stat_res): 
Monte Carlo simulation using a static approach. It simulates only the final index level at maturity and calculates the option value.

# Dynamic Simulation (dyna_res): 
Monte Carlo simulation using a dynamic approach. It simulates the entire path of the index level over the time horizon and calculates the option value based on the path.

# Analytical Solution (anal_res): 
Calculation of option values using the analytical solution provided by the Black-Scholes-Merton model.

```python
def bsm_call_value(S0, K, T, r, sigma):
    ''' Valuation of European call option in BSM model.
    Analytical formula.

    Parameters
    ==========
    S0: float
        initial stock/index level
    K: float
        strike price
    T: float
        maturity date (in year fractions)
    r: float
        constant risk-free short rate
    sigma: float
        volatility factor in diffusion term

    Returns
    =======
    value: float
        present value of the European call option
    '''
    from math import log, sqrt, exp
    from scipy import stats

    S0 = float(S0)
    d1 = (log(S0 / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * sqrt(T))
    d2 = (log(S0 / K) + (r - 0.5 * sigma ** 2) * T) / (sigma * sqrt(T))
    # stats.norm.cdf --> cumulative distribution function
    #                    for normal distribution
    value = (S0 * stats.norm.cdf(d1, 0.0, 1.0) -
             K * exp(-r * T) * stats.norm.cdf(d2, 0.0, 1.0))
    return value
```

```python
# Import the Black-Scholes-Merton call option valuation function

# Initialize empty lists to store results
stat_res = []  # List for results using static simulation
dyna_res = []  # List for results using dynamic simulation
anal_res = []  # List for results using analytical (BSM) approach

# Generate an array of strike prices from 80 to 120 in steps of 5
k_list = np.arange(80., 120.1, 5.)  # Strike prices list

# Set a random seed for reproducibility
np.random.seed(100)

# Loop over each strike price and calculate option values
for K in k_list:
    # Calculate option value using static simulation
    stat_res.append(gbm_mcs_stat(K))  # Append result to stat_res list
    
    # Calculate option value using dynamic simulation
    dyna_res.append(gbm_mcs_dyna(K))  # Append result to dyna_res list
    
    # Calculate option value using the Black-Scholes-Merton (analytical) formula
    anal_res.append(bsm_call_value(S0, K, T, r, sigma))  # Append result to anal_res list

# Convert the result lists to NumPy arrays
stat_res = np.array(stat_res)
dyna_res = np.array(dyna_res)
anal_res = np.array(anal_res)
```

# American Options 

The valuation of American options presents a more complex scenario compared to European options. In the case of American options, determining a fair value involves solving an optimal stopping problem. Equation 12-12 outlines the formulation for valuing an American option as this specific type of problem. It's important to note that the formulation is adapted for a discrete time grid, aligning it with numerical simulation methods.

In a nuanced perspective, it is more accurate to refer to this valuation as an option value given Bermudan exercise. This term acknowledges the discrete time grid used for simulation and the associated complexities involved in determining the optimal exercise strategy over time. As the time intervals between exercises approach zero length, the value of the Bermudan option converges towards that of the American option. This convergence reflects the idea that, in the limit of infinitesimally small time intervals, the Bermudan option behaves similarly to an American option.

```python
import numpy as np
import math

def gbm_mcs_amer(K, option='call'):
    ''' 
    Valuation of American option in Black-Scholes-Merton
    by Monte Carlo simulation by LSM algorithm

    Parameters
    ==========
    K: float
        (positive) strike price of the option
    option: string
        type of the option to be valued ('call', 'put')

    Returns
    =======
    C0: float
        estimated present value of American call option
    '''
    # Calculate the time interval and discount factor
    dt = T / M
    df = math.exp(-r * dt)
    
    # Initialize the array to store the index level paths
    S = np.zeros((M + 1, I))
    S[0] = S0
    
    # Generate random numbers
    sn = gen_sn(M, I)
    
    # Simulate index level paths
    for t in range(1, M + 1):
        S[t] = S[t - 1] * np.exp((r - 0.5 * sigma ** 2) * dt + sigma * math.sqrt(dt) * sn[t])
    
    # Calculate payoff based on the type of option
    if option == 'call':
        h = np.maximum(S - K, 0)
    else:
        h = np.maximum(K - S, 0)
    
    # Initialize the array to store the option values
    V = np.copy(h)
    
    # Apply the LSM algorithm
    for t in range(M - 1, 0, -1):
        reg = np.polyfit(S[t], V[t + 1] * df, 7)
        C = np.polyval(reg, S[t])
        V[t] = np.where(C > h[t], V[t + 1] * df, h[t])
    
    # Calculate MCS estimator
    C0 = df * np.mean(V[1])
    
    return C0

# Call the function with a strike price of 110 for a call option
print(gbm_mcs_amer(110., option='call'))  # Output: 7.721705606305352

# Call the function with a strike price of 110 for a put option
print(gbm_mcs_amer(110., option='put'))  # Output: 13.609997625418051
```


The statement means that, in the context of options pricing, the European option value is considered a lower bound for the American option's value. This is because American options provide the holder with the flexibility to exercise the option at any time before or at expiration, while European options can only be exercised at expiration.

The "early exercise premium" refers to the additional value that an American option holds over its European counterpart due to the early exercise feature. The premium represents the advantage of being able to exercise the option early, potentially capturing favorable market conditions.

```python

# Initialize lists to store the results
euro_res = []
amer_res = []

# Define a list of strike prices
k_list = np.arange(80., 120.1, 5.)

# Calculate the present value of European and American put options for each strike price
for K in k_list:
    euro_res.append(gbm_mcs_dyna(K, 'put'))
    amer_res.append(gbm_mcs_amer(K, 'put'))

# Convert the lists to numpy arrays
euro_res = np.array(euro_res)
amer_res = np.array(amer_res)
```
