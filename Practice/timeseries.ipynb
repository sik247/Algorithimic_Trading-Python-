# Resampling

```python
data.resample('1w', label='right').last().head()
data.resample('1m', label='right').last().head()
```

This statement is highlighting an important consideration when working with time-series data, especially in a financial context, and it relates to the concept of resampling.

In pandas, resampling involves changing the frequency of the data, either downsampling (reducing the frequency) or upsampling (increasing the frequency). The default behavior for many resampling operations in pandas is to use the left label (or index value) of the interval to represent the data in the resampled series. However, in a financial context, it is often preferred to use the right label (or index value) and, in general, the last available data point in the interval.

Here's why this is important:

Consistency with Financial Data:

Financial data, such as stock prices or economic indicators, is often recorded at the end of a specific time period (e.g., end of the trading day). In financial analysis, it's crucial to use the last available data point within a given interval to avoid introducing biases.
Avoiding Foresight Bias:

Foresight bias occurs when future information is included in the analysis at a given point in time, which wouldn't have been known at that specific time. Using the left label might introduce foresight bias because it would involve information that wasn't available until after the end of the time period.
To ensure financial consistency and avoid introducing biases, it is recommended to explicitly specify the closed parameter when using resampling methods in pandas. For example, setting closed='right' ensures that the right label (last available data point) of the interval is used.

# Rollint Statistics

```python
symb = 'AAPL.O'
data = pd.DataFrame(data[symb]).dropna()
data.tail()
```

```python
window = 20  

data['min'] = data[symb].rolling(window=window).min()  
data['mean'] = data[symb].rolling(window=window).mean()  
data['std'] = data[symb].rolling(window=window).std()  
data['median'] = data[symb].rolling(window=window).median()  
data['max'] = data[symb].rolling(window=window).max()  
data['ewma'] = data[symb].ewm(halflife=0.5, min_periods=window).mean()  

data.dropna().head()
```

```python
ax = data[['min', 'mean', 'max']].iloc[-200:].plot(
              figsize=(10, 6), style=['g--', 'r--', 'g--'], lw=0.8) 
data[symb].iloc[-200:].plot(ax=ax, lw=2.0);
```

Rolling statistics are a major tool in the so-called technical analysis of stocks, as compared to the fundamental analysis which focuses, for instance, on financial reports and the strategic positions of the company whose stock is being analyzed.

A decades-old trading strategy based on technical analysis is using two simple moving averages (SMAs). The idea is that the trader should go long on a stock (or financial instrument in general) when the shorter-term SMA is above the longer-term SMA and should go short when the opposite holds true. The concepts can be made precise with pandas and the capabilities of the DataFrame object.

Rolling statistics are generally only calculated when there is enough data given the window parameter specification. As Figure 8-6 shows, the SMA time series only start at the day for which there is enough data given the specific parameterization:

# Simple Moving Average

```python
data['SMA1'] = data[symb].rolling(window=42).mean()
data['SMA2']= data[symb].rolling(window=252).mean()
data[[symb, 'SMA1', 'SMA2']].tail()
```

```python
data[[symb, 'SMA1', 'SMA2']].plot(figsize=(10, 6));
```

```python
data.dropna(inplace=True)

data['positions'] = np.where(data['SMA1'] > data['SMA2'], 1, -1)

ax = data[[symb, 'SMA1', 'SMA2', 'positions']].plot(figsize=(10, 6), secondary_y='positions')
ax.get_legend().set_bbox_to_anchor((0.25, 0.85));
```

# OLS Regression

```python
reg = np.polyfit(rets['.SPX'], rets['.VIX'], deg=1)
```
